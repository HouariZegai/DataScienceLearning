```python
# import libraries
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
df = DataFrame({'X' : [1, 2, 3, 4, 5, 6],
                'Y' : [11, 20, 3, 2, -5, 7]})
df
```

### Line chart (plot)

```python
plt.plot('X', 'Y', data=df) # or: plt.plot(df['X'], df['Y'])
```

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
```

```python
x1, y1 = [1, 2, 3], [1, 5, 3]
x2, y2 = [3, 2, 1], [6, 7, 2]
x3, y3 = [1.5, 2, 3], [9, 4, 3]
# make line chart
plt.plot(x1, y1, label='C')
plt.plot(x2, y2, label='Java')
plt.plot(x3, y3, label='JavaScript')

plt.legend()
plt.title('Programming language')
```

### Bar chart & stacked bar chart (bar)

```python
x, y = [1, 3, 5, 7], [7.5, 3, 6, 0.9]
a, b = [1, 2, 4, 6], [4, 2, 2, 1]
# make bar chart
plt.bar(x, y, label='Algebra', color='#2196f3')
plt.bar(a, b, label='Algorithm', color="y") # y = yellow
plt.title('Subjects')
plt.legend()
```

### Histogram chart (hist)

```python
data = np.random.randn(10000)
# histogram
plt.hist(data, 20, facecolor='#008000')
```

### Scatter chart (scatter)

```python
x = [1, 2, 3, 4, 5, 6]
y = [5, 3, 4, 2, 9, 7]

plt.scatter(x, y, label='scat', color='#219600', s=80, marker="*")
plt.scatter(a, b, label='scat', color='#dd0000', s=80, marker="+")
plt.title('Scatter Title')
plt.legend()
plt.grid()
```

### Area chart (stackplot)

```python
employees = ['Mohammed', 'ALi', 'Mostapha', 'Houari']
# list of tasks
t_a = [20, 15, 25, 30]
t_b = [40, 10, 25, 30]
t_c = [20, 45, 25, 15]
t_d = [20, 30, 25, 25]

plt.plot([], [], color='#333333', label='Task_a', linewidth=5)
plt.plot([], [], color='g', label='Task_b', linewidth=5)
plt.plot([], [], color='#2196f3', label='Task_c', linewidth=5)
plt.plot([], [], color='wheat', label='Task_d', linewidth=5)
# make area chart
plt.stackplot(employees, t_a, t_b, t_c, t_d, colors=['#333333', 'g', '#2196f3', 'wheat'])
plt.legend()
plt.xlabel('Employee')
plt.ylabel('Finished task %')
plt.title('Employee analyse job')
```

### Pie chart (pie)

```python
values = [35, 20, 30, 15]
prog_lang = ['Java', 'C++', 'Python', 'JS']
cols = ['#2196f3', '#008800', '#dd4400', 'y'] # colors

plt.pie(values,
        labels = prog_lang,
        colors = cols,
        startangle = 120,
        explode = (0.09, 0, 0, 0), # change position: (   )
        autopct = '%1.1f%%' # make value + format inside the pie
       )
plt.legend()
plt.title("Most popular Programming Language")
```
